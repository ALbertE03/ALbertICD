```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



with open ("datos.json",'r') as gr:# abrir el json
    dt=gr.read()
granma=json.loads(dt) 

listaños=granma['años']#lista de los años del 85-22



#saldos migratorios internos
smi=granma["SMI_general"]

#tasas migratorias internas
tmi=granma["TMI_general"]

#tasas migratorias externas
tme=granma["TME_general"]

#saldo migratorio externo
sme=granma["SME_general"]


#data frame 
columns= pd.DataFrame([smi,sme,tmi,tme],columns=listaños) #en los [] se necesita poner un lista de 38 elementos

# plt.savefig("diagrama de lineas.png") esto crea un foto.png con el nombre del parametro

plt.fill_between(listaños,sme)# diagrama de areas del saldo migratorio externo
plt.show()
plt.fill_between(listaños,smi) #diagrama de areas del saldo migratorio interno
plt.show()
plt.fill_between(listaños,tmi) # diagrama de areas de la tasa migratoria interna
plt.show()
plt.fill_between(listaños,tme) # diagrama de areas de la tasa migratoria extera
plt.show()
```
